#### Solution 

```python
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        arr3=nums1+nums2
        arr3.sort()
        m=len(arr3)
        if m%2==1:
            return arr3[m//2]
        else:
            mid=m//2
            return (arr3[mid-1]+arr3[mid])/2
        
```
